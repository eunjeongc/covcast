```python
df2 = pd.read_excel(r"C:\Users\gebruikerr\OneDrive\Desktop\population.xlsx") 
# Filter the DataFrame]
df2 = df2.drop(columns=df2.columns[0])
df2 = df2.dropna()
# Keep only the first two columns
df2= df2.iloc[:, :2]
df2 = df2[['Population Estimate \n(as of July 1)']]
# Repeat each row 17 times
df2 = df2.loc[df2.index.repeat(17)].reset_index(drop=True)

display(df2)
```

```python
import pandas as pd

# Assuming df is your original DataFrame and df_second_column is the DataFrame with only the second column

# Reset the index of df_second_column
df2.reset_index(drop=True, inplace=True)

# Concatenate the DataFrames along axis 1 (columns)
df_concatenated = pd.concat([df, df2], axis=1)

df_concatenated = df_concatenated.dropna()
# Display the updated DataFrame
display(df_concatenated)
```

```python
df_concatenated['inc_case_per_state'] = df_concatenated['inc_case'] / df_concatenated['Population Estimate \n(as of July 1)']

display(df_concatenated)
```
